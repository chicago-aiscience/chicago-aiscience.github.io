```python
import json
from pathlib import Path

BASE_DIR = Path(".")

print('done')
print(BASE_DIR.cwd())
```

```python
with open(BASE_DIR / "configs" / "fellows.json") as f:
    fellows = json.load(f)

OUTDIR = BASE_DIR / "apps" / "site" / "src" / "content" / "fellows"
OUTDIR.mkdir(exist_ok=True)
print(OUTDIR)

for fellow in fellows:
    cohort = fellow["cohort"]
    name = fellow["name"]

    name_parts = name.lower().split()
    first_initial = name_parts[0][0]
    lastname = name_parts[-1]

    filename = f"{cohort}-{first_initial}-{lastname}.md"
    content = f"""---
metaName: {name}
idCohort: "{cohort}"
idGithub: {fellow.get('githubUser', [])}
metaImageUrl: {fellow.get('imageUrl', [])}
metaTitle: {name}
isSchmidtFellow: true
---
"""
    
    with open(OUTDIR / filename, "w") as f:
        f.write(content)
```
